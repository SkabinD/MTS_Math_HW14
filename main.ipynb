# Задание 1

На лекции мы получили два вида формулы для F-статистики:

$$ F = \frac{(RSS_{res} - RSS_{unres}) / q}{RSS_{unres}/(n - k - 1)} $$

$$ F = \frac{(R_{unres}^2 - R_{res}^2) / q}{(1 - R_{unres}^2)/(n - k - 1)} $$

Необходимо продемонстрировать вывод F-статистики из формулы 1 в формулу 2

## Решение

В усеченной модели, в отличии от полной модели, отсутствуют некоторые регрессоры. В таком случае $RSS_{unres} \neq RSS_{res}$.

По определению $TSS = ESS + RSS = \sum_{i=1}^n(y_i - \bar{y})$ и не зависит от количества регрессоров. Тогда формулу 1 можно переписать в виде:

$$ F = \frac{(TSS - ESS_{res} - TSS + ESS_{unres}) / q}{(TSS - ESS_{unres})/(n - k - 1)} = \frac{(ESS_{unres} - ESS_{res}) / q}{(TSS - ESS_{unres})/(n - k - 1)} $$

Домножим числитель и знаменатель на $1 / TSS$ и с учетом того что $R^2 = ESS / TSS$:

$$ F = \frac{\frac{ESS_{unres} - ESS_{res}}{TSS} \frac{1}{q}}{\frac{TSS - ESS_{unres}}{TSS} \frac{1}{n - k - 1}} = \frac{(R_{unres}^2 - R_{res}^2)/q}{(1 - R_{unres}^2) /(n - k - 1)}.$$

## Подготовка датасета

```python
PARAMS = {
    'x1_mean': 100.0,
    'x2_mean': 50.0,
    'x3_mean': 10.0,
    
    'x1_std': 25.0,
    'x2_std': 50.0,
    'x3_std': 10.0,
    
    'corr12': 0.0,
    'corr13': 0.0,
    'corr23': 0.8,

    'e_mean': 0,
    'e_std': 50,

    'beta0': 20,
    'beta1': 5,
    'beta2': 10,
    'beta3': 20
}

SIZE = 1000
SEED = 1337
```

```python
def gen_data_hw(params, size, seed):
    np.random.seed(seed)

    means = [params['x1_mean'], params['x2_mean'], params['x3_mean']]

    var1 = params['x1_std'] ** 2
    var2 = params['x2_std'] ** 2
    var3 = params['x3_std'] ** 2

    cov12 = params['corr12'] * params['x1_std'] * params['x2_std']
    cov13 = params['corr13'] * params['x1_std'] * params['x3_std']
    cov23 = params['corr23'] * params['x2_std'] * params['x3_std']

    X = np.random.multivariate_normal(
            mean=means,
            cov=[[var1, cov12, cov13],
                 [cov12, var2, cov23],
                 [cov13, cov23, var3]],
            size=size
        )
    
    e = np.random.normal(loc=params['e_mean'], scale=params['e_std'], size=size)
    
    y =   params['beta0'] \
        + params['beta1'] * X[:, 0] \
        + params['beta2'] * X[:, 1] \
        + params['beta3'] * X[:, 2] \
        + e

    dataset = pd.DataFrame({'x1': X[:, 0], 'x2': X[:, 1], 'x3': X[:, 2],'e': e,'y': y})

    return dataset

data = gen_data_hw(PARAMS, SIZE, SEED)

data.describe()
```

#### b. Проверка работы модели на разном N

```python
N = [30 + i for i in range(971)]

beta1 = []
beta2 = []
beta1_se = []
beta2_se = []
beta1_l = []
beta1_r = []
beta2_l = []
beta2_r = []

for n in N:
    data = gen_data_hw(PARAMS, n, SEED)

    _, model = train_model(dataset=data, target='y', feature_names=['x1', 'x2'], 
                                show_results=False, pairwise=False, return_norm_tests=False)
    
    beta1.append(model.params['x1'])
    beta2.append(model.params['x2'])
    beta1_se.append(model.bse['x1'])
    beta2_se.append(model.bse['x2'])
    
    l, r = model.conf_int(alpha=0.05).loc['x1']
    
    beta1_l.append(l)
    beta1_r.append(r)
    
    l, r = model.conf_int(alpha=0.05).loc['x2']
    
    beta2_l.append(l)
    beta2_r.append(r)

size_dependency_df = pd.DataFrame({'N': N,
                                   'beta1': beta1,
                                   'beta2': beta2,
                                   'beta1_se': beta1_se,
                                   'beta2_se': beta2_se,
                                   'beta1_l': beta1_l,
                                   'beta1_r': beta1_r,
                                   'beta2_l': beta2_l,
                                   'beta2_r': beta2_r})
```

```python
size_dependency_df
```

```python
# Графики зависимости стандартной ошибки от количества экземпляров в выборке

plt.title('Зависимость стандартной ошибки от N')
plt.plot(size_dependency_df['N'], size_dependency_df['beta1_se'], label='beta1')
plt.plot(size_dependency_df['N'], size_dependency_df['beta2_se'], label='beta2')
plt.legend()
plt.xlabel('N')
plt.ylabel('std err')
plt.grid()
plt.show()
```

#### c. График сходимости оценок коэффициентов

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

fig.suptitle('Сходимость оценок коэффициентов')

axes[0].plot(size_dependency_df['N'], size_dependency_df['beta1'], label='beta1')
axes[0].fill_between(x=size_dependency_df['N'], y1=size_dependency_df['beta1_l'], y2=size_dependency_df['beta1_r'], label='Доверительный интервал', color='red')
axes[0].axhline(PARAMS['beta1'], color='black', linestyle='dashed', label='Истинное значение')

axes[1].plot(size_dependency_df['N'], size_dependency_df['beta2'], label='beta2')
axes[1].fill_between(x=size_dependency_df['N'], y1=size_dependency_df['beta2_l'], y2=size_dependency_df['beta2_r'], label='Доверительный интервал', color='red')
axes[1].axhline(PARAMS['beta2'], color='black', linestyle='dashed', label='Истинное значение')

for ax in axes:
    ax.legend()
    ax.set_xlabel('N')
    ax.grid()

plt.show()
```

На основе полученных графиков можно сделать выводы о том, что сходимость оценок коэффициентов достигается при $N \approx 600$, дальнейший прирост точности становится уже не таким значительным. Доверительные интервалы содержат истинные значения для обоих коэффициентов и распологаются симметрично относитльно них, что говорит о несмещенности получаемых оценок, а уменьшение ошибки при увеличении размера выборки говорит о состоятельности таких оценок.
